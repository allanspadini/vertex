```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = 'morango_teste.png'
image = cv2.imread(image_path)

# Convert BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Given bounding boxes for ripe strawberries (as xmin, ymin, xmax, ymax)
bboxes = [
    [0.2923565, 0.3833163, 0.49518442, 0.6222993],
    [0.46259013, 0.5574023, 0.5006691, 0.6442665],
    [0.3799183, 0.46671477, 0.50627977, 0.63586146],
    [0.4595869, 0.59002733, 0.28562254, 0.4565382],
    [0.60172987, 0.7236054, 0.2793683, 0.4407453],
    [0.2984194, 0.3859305, 0.32159588, 0.44795316],
    [0.26325116, 0.31194678, 0.54023993, 0.62218356],
    [0.61243534, 0.7506516, 0.5909766, 0.77752316],
    [0.3118624, 0.38402164, 0.18442589, 0.30913663],
    [0.9440251, 0.9972117, 0.46084622, 0.5341813],
    [0.12076803, 0.17719719, 0.49393657, 0.5679644],
    [0.031458333, 0.08729285, 0.3774913, 0.44935593],
    [6.9568865E-4, 0.013312342, 0.36920062, 0.45814323],
    [0.30715618, 0.36301693, 0.6076706, 0.67493284],
    [0.51717377, 0.6456874, 0.29113665, 0.45596913],
    [0.08865748, 0.12131098, 0.41696858, 0.46504644],
    [2.2936147E-4, 0.027139274, 0.38147208, 0.49431574],
    [0.0015325453, 0.013665636, 0.43424246, 0.5116756],
    [0.47671542, 0.5361943, 0.4484692, 0.52331585],
    [0.0, 0.006482536, 0.3262058, 0.43157518],
    [0.11786988, 0.15242389, 0.43647575, 0.48944986],
    [0.29924774, 0.38877258, 0.36631402, 0.5084448],
    [0.06565795, 0.09556162, 0.39711642, 0.44768193],
    [5.336171E-4, 0.023129959, 0.43402055, 0.4951808],
    [3.790208E-4, 0.011480609, 0.320353, 0.44187993]
]

# Function to draw bounding boxes on the image
def draw_boxes(image, boxes):
    h, w, _ = image.shape  # Get the dimensions of the image
    for box in boxes:
        # Convert from relative to absolute coordinates
        xmin, ymin, xmax, ymax = box
        pt1 = (int(xmin * w), int(ymin * h))
        pt2 = (int(xmax * w), int(ymax * h))

        # Draw the rectangle
        image = cv2.rectangle(image, pt1, pt2, (0, 255, 0), 2)
    return image

# Draw the bounding boxes on the image
boxed_image = draw_boxes(image, bboxes)

# Display the image with bounding boxes
plt.figure(figsize=(10,10))
plt.imshow(boxed_image)
plt.axis('off') # Hide the axis
plt.show()
```
